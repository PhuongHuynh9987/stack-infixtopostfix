```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
import random
```

```python
Path.cwd()
taylor = pd.read_csv('/Users/macmall/Desktop/NLP/taylor_swift_lyrics.csv',encoding='latin1')
taylor
```

```python
emotion  = pd.read_csv('/Users/macmall/Desktop/NLP/Emotion_final.csv',encoding='latin1')
emotion
```

```python
vec = TfidfVectorizer(stop_words = 'english') # model that computes tf idf 
X = vec.fit_transform(taylor['lyric'])               # apply model to our data (quotes)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out()) # display as DataFrame
df
```

```python
df1 = df.sum() > 4.29
df1 = df.columns[df1].tolist()
```

```python
update_df = pd.DataFrame()
for i in df:
    if i in df1:
        update_df = pd.concat([update_df,df[i]],axis =1)
update_df
```

```python
list_of_word = [0]*len(taylor)
list_of_word = pd.Series(list_of_word)

for i in update_df:
    j = update_df[i]
#     print(i)
    list_of_word = list_of_word.where(update_df[i] != update_df[i].max(), i)
list_of_word

taylor_update = taylor.copy()
taylor_update['words'] = list_of_word
# print(list_of_word.tolist())
```

```python
count = taylor_update.groupby(['year','words']).words.count()
count
```

```python
X_train, X_test, y_train, y_test = train_test_split(taylor['lyric'],taylor['year'], test_size=0.2)
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, y_train)

pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib as mpl
import matplotlib.pyplot as plt

year = taylor['year'].unique()

mat = confusion_matrix(y_test, pred) # create the confusion matrix

fig, ax = plt.subplots(figsize = (6,6))
ax.imshow(mat, cmap='OrRd')     # making the heat-map image
ax.set_xticks(np.arange(len(year)))
ax.set_xticklabels(year, rotation = 90,size = 7) # label ticks with newsgroups
 
ax.set_yticks(np.arange(len(year)))
ax.set_yticklabels(year)

for i in range(len(year)):                # Loop over data dimensions and create text annotations:
    for j in range(len(year)):
        text = ax.text(j, i, mat[i, j], ha="center", va="center", color="k");
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test, pred)
```
